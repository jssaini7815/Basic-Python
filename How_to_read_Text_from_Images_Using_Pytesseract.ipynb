```
import pytesseract
from google.colab.patches import cv2_imshow
```

```
img=cv2.imread('/content/drive/MyDrive/Hare_krsna.jpg')
cv2_imshow(img)
cv2.waitKey(0) #press any key
cv2.destroyAllWindows() 
```

```
text=pytesseract.image_to_string(img)
print(text)
```
